```python
import numpy as np
import matplotlib.pyplot as plt


N = 2
width = 0.35 

comp_time_ape = [ 1822, 211 ]
comp_time_mee = [ 705, 225]

fig, ax = plt.subplots(
    figsize=[8,5]
)

ax.set_xlim([-0.50,2])
ax.set_ylim([0,1850])

ind = np.arange(1.5)    # the x locations for the groups

         # the width of the bars
p1 = ax.bar(
    ind, 
    comp_time_ape, 
    width, 
    bottom=0,
    color='b'
)

p2 = ax.bar(
    ind + width, 
    comp_time_mee, 
    width, 
    bottom=0,
    color='r',
    alpha=1
)

ax.set_title('Comaprison of computation time : APE and MEE', fontsize=16)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Connect 4', 'Cover Type'))
# ax.set_yticklabels([0,200,500,1000,1500,1800])
plt.grid(linestyle='-', linewidth='0.3', color='c',alpha=0.75)
plt.tick_params(
    axis='both', 
    which='major',
    labelsize=16
)
ax.legend(
    (p1[0], p2[0]), 
    ('APE', 'MEE'),
    loc='upper right',
    fontsize=16
)
plt.tight_layout()
# ax.autoscale_view()
plt.savefig('computation_time_1.png')
plt.show()
```
